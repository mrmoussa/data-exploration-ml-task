```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
titanic_data = pd.read_csv("train.csv", index_col='PassengerId')
```

```python
#The graph shows that the number of male passengers was higher than female passengers.
plt.figure(figsize=(10,6))
plt.title('The distribution of Male/Female passengers in Titanic')
sns.countplot(x='Sex', hue='Sex', data=titanic_data)
```

```python
#Overall distribution of survived people
#It is clear from the graph below that more than 50% of passenger were victims of Titanic sunk
plt.figure(figsize=(10,6))
plt.title('The graph showing the number for survived and deceased people')
sns.countplot(x='Survived', hue='Survived', data=titanic_data)
```

```python
#The distribution of survived people considering their gender. 
# It is true that in this tragedy more female were taken to boats to save their lives
plt.figure(figsize=(10,6))
plt.title('The distribution of Male/Female passengers survival')
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
```

```python
#The distribution of people survived among the ticket classes. It is clear from the graph people from the first class had
#higher chances of survival. Unfortunately, most of the people of from Class 3 lost their lives that day.
plt.figure(figsize=(10,6))
plt.title('The distribution of passengers survival across Classes 1, 2, 3')
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
```

```python
#According to this graph there were approximately 90% people with at most 1 sibling/spouse aboard of Titanic.
plt.figure(figsize=(10,6))
plt.title('The number of siblings/spouse being aboard on Titanic')
sns.countplot(x='SibSp', hue='Survived', data=titanic_data)
```

```python
#The graph tells us that many people were without parents/children that day and approximately 8000 people were with 1 or both parents
#aboard that day.
plt.figure(figsize=(10,6))
plt.title('The number of parents/children being aboard on Titanic')
sns.countplot(x='Parch', hue='Survived', data=titanic_data)
```

```python
#From the graph below it's clear that the majority of people(50000) were embarked at Southampton port, and the rest approximately 18000
#from Cherbourg port and  the rest (2000-3000) were from Queenstown.
plt.figure(figsize=(10,6))
plt.title('The number of people embarked from different ports')
sns.countplot(x='Embarked', hue='Survived', data=titanic_data)
```

```python
#The graph below shows the distribution of people survived and become victims according to their Age.
#The graphs shows that majority of people who were deceased are between ages 20-40
f, ax = plt.subplots(1, 1,figsize=(10,10))

sns.distplot(ax= ax,a = titanic_data[titanic_data['Survived']==1]['Age'], color="blue", label="Survived")
sns.distplot(ax=ax,a = titanic_data[titanic_data['Survived']==0]['Age'], color="red", label="Deceased")

plt.legend(labels=['Survived', 'Deseaced'])
ax.set_xlabel("Age")
plt.show()
```

```python
#It's clear that people who were older were survived rather than younger people.
plt.figure(figsize=(8,8))
plt.title('Boxplot graph between age and survived columns')
sns.boxplot(x='Survived', y='Age', data=titanic_data)
titanic_data['Age'].corr(titanic_data['Survived'])
```
