```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
titanic_data = pd.read_csv("train.csv", index_col='PassengerId')
```

```python
titanic_data.head()
```

```python
plt.figure(figsize=(10,6))
plt.title('Titanic Age groups')
sns.barplot(x=titanic_data.index, y=titanic_data['Age'])
```

```python
plt.figure(figsize=(10, 8))
sns.distplot(a=titanic_data['Age'], kde=False)
```
